```python
#sleep_in 
def sleep_in(weekday, vacation):
 if weekday == False or vacation == True:
   return True
 else:
  return False
```

```python
#monkey_trouble 
def monkey_trouble(a_smile, b_smile):
  if a_smile == b_smile:
    return True
  else:
   return False
```

```python
#sum_double 
def sum_double(a, b):
  if a==b:
    return 2*(a+b)
  else:
    return a+b
```

```python
#diff21 
def diff21(n):
  if n > 21:
    return 2*(n-21)
  else:
    return 21-n
```

```python
#parrot_trouble 
def parrot_trouble(talking, hour):
  if talking == True and (hour <7 or hour >20):
    return True
  else:
   return False
```

```python
# makes10
def makes10(a, b):
  if a == 10 or b == 10:
    return True
  else:
    if a+b == 10:
      return True
    else:
      return False
```

```python
#near_hundred 
def near_hundred(n):
  if abs(100-n) <= 10 or abs (200-n) <= 10:
    return True
  else:
    return False
```

```python
#pos_neg 
def pos_neg(a, b, negative):
 if negative == False:
   if (a<0 and b>0) or (a>0 and b<0):
    return True
   else:
     return False
 if negative==True:
   if a<0 and b<0:
    return True
   else:
    return False
```

```python
#not_string 
def not_string(str):
  if str[:3]=='not':
    return str
  else:
    return 'not ' + str
```

```python
#missing_char
def missing_char(str, n):
 if n <= len(str)-1:
   str= str[0:n]+str[n+1:len(str)]
   return str
```

```python
#front_back 
def front_back(str):
 if len(str) <= 1:
    return str
 start= str[len(str)-1]
 mid= str[1:len(str)-1]
 end=str[0]
 return start+mid+end
```

```python
#cigar_party 
def cigar_party(cigars, is_weekend):
 if is_weekend==True:
   if cigars>=40:
     return True
   else:
     return False
 elif cigars>=40 and cigars<=60:
     return True
 else:
     return False
```

```python
#date_fashion 
def date_fashion(you, date):
 if you<=2 or date<=2:
     return 0
 elif you>=8 or date>=8:
  return 2
 else:
  return 1
```

```python
#squirrel_play 
def squirrel_play(temp, is_summer):
  if is_summer==True:
    if temp>=60 and temp<=100:
       return True
    else:
       return False
  elif temp>=60 and temp<=90:
       return True
  else:
    return False
```

```python
#caught_speeding 
def caught_speeding(speed, is_birthday):
 if is_birthday==True:
   speed=speed-5
 if speed<=60:
   return 0
 elif speed>=61 and speed<=80:
   return 1
 else:
  return 2
```

```python
#sorta_sum 
def sorta_sum(a, b):
 if a+b>=10 and a+b<=19:
   return 20
 else:
   return a+b
```

```python
#alarm_clock 
def alarm_clock(day, vacation):
 alarm1='7:00'
 alarm2='10:00'
 noalarm='off'
 if vacation==True:
   if day>=1 and day<=5:
     return alarm2
   else:
     return noalarm
 else:
   if day>=1 and day<=5:
     return alarm1
   else:
     return alarm2
```

```python
#love6 
def love6(a, b):
 if (a==6 or b==6) or (a+b==6) or (abs(a-b)==6):
   return True
 else:
   return False
```

```python
#in1to10
def in1to10(n, outside_mode):
 if outside_mode==True:
   if n<=1 or n>=10:
     return True
   else:
     return False
 elif n>=1 and n<=10:
    return True
 else:
   return False
```
